```python
import pandas as pd

# Carga de datasets
df_all = pd.read_csv('../../../dataset/cleaned_filtered/df_filtrado_limpio.csv', sep=',')

df_all.head()
```

```python
# guarda ID + DB_SAL_CUE_ + DB_CLI_ + target en un csv
vars_seleccionadas = [col for col in df_all.columns if col.startswith('DB_SAL_CUE_') or col.startswith('DB_CLI_') or col.startswith('DB_SAL_CRE_')]
df_sald_cue_clientes_target = df_all[['ID'] + vars_seleccionadas + ['target']].copy()
df_sald_cue_clientes_target.to_csv('../../../dataset/cleaned_filtered/df_sald_cue_clientes_target.csv', index=False, encoding='utf-8')


# guarda ID + DB_SAL_CUE_ + DB_CLI_ + DB_SAL_CRE_ + DB_CONS_ + target en un csv
vars_seleccionadas = [col for col in df_all.columns if col.startswith('DB_SAL_CUE_') or col.startswith('DB_CLI_') or col.startswith('DB_SAL_CRE_') or col.startswith('DB_CONS_')]
df_sald_cue_clientes_cre_cons_target = df_all[['ID'] + vars_seleccionadas + ['target']].copy()
df_sald_cue_clientes_cre_cons_target.to_csv('../../../dataset/cleaned_filtered/df_sald_cue_clientes_cre_cons_target.csv', index=False, encoding='utf-8')    
```
